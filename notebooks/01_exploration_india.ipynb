```python
import os
import pandas as pd

BASE_DIR = os.path.abspath(os.path.join(os.getcwd(), ".."))
print("BASE_DIR =", BASE_DIR)

file_path = os.path.join(BASE_DIR, "data", "raw", "districtwise-ipc-crimes-2017-onwards.csv")

print("file exists?", os.path.exists(file_path))

df = pd.read_csv(file_path)
df.head()
```

```python
import pandas as pd

# columns that are not crimes
id_cols = [
    "id", "year", "state_name", "state_code",
    "district_name", "district_code", "registration_circles"
]

crime_cols = [c for c in df.columns if c not in id_cols]

df_long = df.melt(
    id_vars=["year", "state_name", "district_name"],
    value_vars=crime_cols,
    var_name="crime_type",
    value_name="crime_count"
)

# Ensure numeric
df_long["crime_count"] = pd.to_numeric(df_long["crime_count"], errors="coerce").fillna(0)

df_long.head()
```

```python
district_totals = (
    df_long.groupby(["year", "state_name", "district_name"], as_index=False)
    .agg(total_crime=("crime_count", "sum"))
)

district_totals.sort_values("total_crime", ascending=False).head(10)
```

```python
state_totals = (
    df_long.groupby(["year", "state_name"], as_index=False)
    .agg(total_crime=("crime_count", "sum"))
)

state_totals.sort_values(["year", "total_crime"], ascending=[False, False]).head(15)
```

```python
import pandas as pd

master = pd.read_csv("../data/processed/master_crime_long.csv")

# 1) State-wise yearly totals
state_year = (
    master.groupby(["year", "state_name"], as_index=False)
          .agg(total_crime=("crime_count", "sum"))
)

# 2) District-wise yearly totals (hotspot detection)
district_year = (
    master.groupby(["year", "state_name", "district_name"], as_index=False)
          .agg(total_crime=("crime_count", "sum"))
)

# 3) Crime-type distribution per year (category-based)
type_year = (
    master.groupby(["year", "category", "crime_type"], as_index=False)
          .agg(total_crime=("crime_count", "sum"))
)

print("✅ state_year:", state_year.shape)
print("✅ district_year:", district_year.shape)
print("✅ type_year:", type_year.shape)

state_year.head()
```
